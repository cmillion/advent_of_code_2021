```python
def parse_input(input):
    input = input.split()
    moves = np.array(input[0].split(','),dtype='uint32')
    boards = []
    for j in range(int(len(input[1:])/25)):
        board = np.zeros(5*5)
        for i in range(5*5):
            board[i] = int(input[1+i+25*j])
        boards+=[np.reshape(board,[5,5]).tolist()]
    return moves, boards
```

```python
moves, boards = parse_input(example_games)
print(moves)
```

```python
def evaluate_board_states(markers):
    for j,marker in enumerate(markers):
        for i in range(5):
            if np.array(marker)[i,:].sum()==0: # check rows
                return j,marker
            if np.array(marker)[:,i].sum()==0: # check columns
                return j,marker
    return None, None
```

```python
def play_bingo(moves,boards):
    # unmarked entries are 1 so that we can just multiply to mask
    markers = np.ones(np.shape(boards),dtype='uint8')
    for i in moves:
        ix = np.where(np.array(boards,'uint64')==i)
        markers[ix]=0
        board_num,winning_board = evaluate_board_states(markers)
        if board_num:
            return boards[board_num], markers[board_num], i
        #else no board has won; continue playing
```

```python
moves, boards = parse_input(example_games)
board,markers,winning_move = play_bingo(moves,boards)
assert (np.array(board)*np.array(markers)).sum()*winning_move==4512
```
