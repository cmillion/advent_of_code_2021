```python
example_commands = """
forward 5
down 5
forward 8
up 3
down 8
forward 2
"""
```

```python
def reposition(commands):
    position,depth = 0,0
    for command in commands.split('\n'):
        if command=='':
            continue
        direction, impulse = command.split()
        if direction=='forward':
            position+=int(impulse)
        elif direction=='down':
            depth+=int(impulse)
        elif direction=='up':
            depth-=int(impulse)
    return position, depth
```

```python
assert np.product(reposition(example_commands))==150
```

### Part 2

```python
def reposition2(commands):
    position,depth,aim = 0,0,0
    for command in commands.split('\n'):
        if command=='':
            continue
        direction, impulse = command.split()
        if direction=='forward':
            position+=int(impulse)
            depth+=aim*int(impulse)
        elif direction=='down':
            aim+=int(impulse)
        elif direction=='up':
            aim-=int(impulse)
    return position, depth
```

```python
assert np.product(reposition2(example_commands))==900
```
